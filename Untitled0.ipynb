```python
import random

numero_persona = int(input("Dame un número: "))

# La máquina decide par o impar
eleccion_maquina = random.choice(["par", "impar"])

print("Tu número es:", numero_persona)
print("La máquina eligió:", eleccion_maquina)

# Determinar si el número es par o impar
if numero_persona % 2 == 0:
    resultado = "par"
else:
    resultado = "impar"

print("El número es:", resultado)

# Ver quién gana
if resultado == eleccion_maquina:
    print("¡Ha ganado la máquina, MATRIX!")
else:
    print("¡Ha ganado la persona!")
```

```python
alfabeto ='abcdefghijklmnopqrtuvwxyz'
mensaje = input("Introduce el mensaje a cifrar: ")
desplazamiento = 3
resultado = ""

for letra in mensaje:
    if letra.isalpha():
        base = ord('A') if letra.isupper() else ord('a')
        nueva_letra = chr((ord(letra) - base + desplazamiento) % 26 + base)
        resultado += nueva_letra
    else:
        resultado += letra

print("Mensaje cifrado:", resultado)
```

```python
import hashlib
data = input("Dame un texto:")
encode_data = data.encode('utf-8')
md5_hash = hashlib.md5()
md5_hash.update(encode_data)
hex_digest = md5_hash.hexdigest()
print(f"Original string: {data}")
print(f"MD5 hash: {hex_digest}")
```
